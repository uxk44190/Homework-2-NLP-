```python
from collections import Counter, defaultdict

# ----------------------------
# 1) Training corpus (given)
# ----------------------------
corpus = [
    "<s> I love NLP </s>",
    "<s> I love deep learning </s>",
    "<s> deep learning is fun </s>"
]

# ----------------------------
# 2) Build unigram + bigram counts
# ----------------------------
unigram_counts = Counter()
bigram_counts = Counter()

for sent in corpus:
    tokens = sent.split()
    unigram_counts.update(tokens)

    # bigrams
    for i in range(len(tokens) - 1):
        bigram = (tokens[i], tokens[i + 1])
        bigram_counts[bigram] += 1

# ----------------------------
# 3) Bigram probability (MLE)
#    P(w2 | w1) = count(w1,w2) / count(w1)
# ----------------------------
def bigram_prob(w1, w2):
    return bigram_counts[(w1, w2)] / unigram_counts[w1]

# ----------------------------
# 4) Sentence probability function
#    Multiply all bigram probabilities in the sentence
# ----------------------------
def sentence_probability(sentence):
    tokens = sentence.split()
    prob = 1.0

    for i in range(len(tokens) - 1):
        w1, w2 = tokens[i], tokens[i + 1]

        # If a bigram never appeared, probability becomes 0 (MLE)
        if bigram_counts[(w1, w2)] == 0:
            return 0.0

        prob *= bigram_prob(w1, w2)

    return prob

# ----------------------------
# 5) Test on both sentences (given)
# ----------------------------
s1 = "<s> I love NLP </s>"
s2 = "<s> I love deep learning </s>"

p1 = sentence_probability(s1)
p2 = sentence_probability(s2)

print("Unigram counts:", unigram_counts)
print("Bigram counts:", bigram_counts)
print()
print("Sentence 1:", s1)
print("P(S1) =", p1)
print()
print("Sentence 2:", s2)
print("P(S2) =", p2)
print()

# ----------------------------
# 6) Print which sentence is preferred and why
# ----------------------------
if p1 > p2:
    print("Model prefers Sentence 1 because it has higher bigram probability.")
elif p2 > p1:
    print("Model prefers Sentence 2 because it has higher bigram probability.")
else:
    print("Model gives both sentences the same probability.")
```
