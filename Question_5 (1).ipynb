```python
import numpy as np

# Confusion matrix: rows = predicted (system), cols = actual (gold)
cm = np.array([
    [5, 10, 5],   # predicted Cat
    [15, 20, 10], # predicted Dog
    [0, 15, 10]   # predicted Rabbit
])

labels = ["Cat", "Dog", "Rabbit"]
total = cm.sum()

per_class = {}
for i, lab in enumerate(labels):
    TP = cm[i, i]
    FP = cm[i, :].sum() - TP
    FN = cm[:, i].sum() - TP

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0.0
    recall    = TP / (TP + FN) if (TP + FN) > 0 else 0.0

    per_class[lab] = (precision, recall)

# Macro averages
macro_precision = sum(p for p, r in per_class.values()) / len(labels)
macro_recall    = sum(r for p, r in per_class.values()) / len(labels)

# Micro averages
TP_total = np.trace(cm)
micro_precision = TP_total / total
micro_recall = TP_total / total

print("Per-class Precision & Recall:")
for lab in labels:
    p, r = per_class[lab]
    print(f"{lab}: Precision = {p:.4f}, Recall = {r:.4f}")

print("\nMacro-averaged:")
print(f"Macro Precision = {macro_precision:.4f}")
print(f"Macro Recall    = {macro_recall:.4f}")

print("\nMicro-averaged:")
print(f"Micro Precision = {micro_precision:.4f}")
print(f"Micro Recall    = {micro_recall:.4f}")
```
